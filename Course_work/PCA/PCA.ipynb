```python
"""
Patterns:

1

1    1
    01

0       11
0       01
111

1000
1       011
1       001
1       111

        11111
1000        0
1011        0
1001        0
1111        0


"""

def my_spiral_ones(n, direction):
    if n == 1:
        return [[1]]
    
    new_dir = (direction - 1) % 4
    submatrix = my_spiral_ones(n - 1, new_dir)  # Change direction
    matrix = [[0] * n for _ in range(n)]

    if direction == 0:  # Right
        for i in range(n):
            matrix[i][-1] = 1

        for i in range(n - 1):
            for j in range(n - 1):
                matrix[i][j] = submatrix[i][j]
    
    elif direction == 1:  # Down
        for i in range(0, n-1):
            for j in range(1, n):
                matrix[i][j] = submatrix[i][j-1]

        for i in range(n):
            matrix[-1][i] = 1
            
    
    elif direction == 2:  # Left
        for i in range(1, n):
            for j in range(1, n):
                matrix[i][j] = submatrix[i-1][j-1]
        for i in range(n):
            matrix[i][0] = 1
    
    elif direction == 3:  # Up
        for i in range(1, n):
            for j in range(n - 1):
                matrix[i][j] = submatrix[i-1][j]
        for i in range(n):
            matrix[0][i] = 1
    
    return matrix

# Test the function with n=4 and initial direction "R" (0)
n = 9
d = (n - 2)%4
result = my_spiral_ones(n,d)
for row in result:
    for i in row:
        print(i, end = " ")
    print()
```

```python
def is_palindrome(s):
    return s == s[::-1]

def longestPalindrome(s):
    palindromic_substring=[]
    for i in range (len(s)):
        for j in range (i, len(s)):
            sub_string = s[i:j+1]
            if is_palindrome(sub_string):
                if len(palindromic_substring)>len(sub_string):
                    continue
                else:
                    palindromic_substring = sub_string
    return palindromic_substring

print(longestPalindrome("acbcyaababaatu"))
```

```python
import sys

def minCoins(sum, coins, types, table):
    # If the sum is reaching 0, no more coins are needed.
    if(sum == 0):
        return 0
    
    # Searching if memoized
    if table[sum] != -1:
        return table[sum]
    
    res = sys.maxsize

    for i in range(0, types):
        # If value of coin is less than remaining sum, only then we take it and recursively call to check further
        if(coins[i] <= sum):
            sub_res = minCoins(sum - coins[i], coins, types, table)

            if(sub_res != sys.maxsize and sub_res + 1 < res):
                res = sub_res + 1
    
    table[sum] = res
    return res

coins = [9, 8, 7]
sum = 6
table = [-1] * (sum + 1)
ans = minCoins(sum, coins, len(coins), table)

if ans == sys.maxsize:
    ans = -1

print(ans)
```

```python
def longsubstr(string, k):
    map = {}
    res = ""
    i,j = 0, 0

    while j < len(string):
        map[string[j]]=1+map.get(string[j],0)

        if len(map) < k:
            j += 1
        elif len(map) == k:
            if(j - i + 1> len(res)):
                res = string[i:j+1]
            j+=1
        elif len(map) > k:
            while len(map) > k:
                map[string[i]] -= 1
                if map[string[i]]==0:
                    del map[string[i]]
                i+=1
                if(len(map)==k):
                    if(j - i + 1> len(res)):
                        res = string[i:j+1]
            j+=1
    return res

print(longsubstr("abcba", 2))
```

```python
def my_rec_det(arr):
    n = len(arr)
    if n == 1:
        det = arr[0][0]
    else:
        det = 0
        for j in range(n):
            sub_matrix = [[0]*(n-1) for i in range(n-1)]
            for i in range(1,n):
                for k in range(n-1):
                    if k < j:
                        sub_matrix[i-1][k] = arr[i][k]
                    elif k >= j:
                        sub_matrix[i-1][k] = arr[i][k+1]
            det += my_rec_det(sub_matrix) * ((-1)**j) * arr[0][j]
    return det


arr = [[1,1],[0,1]]
det = my_rec_det(arr)
print(det)
```

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

# Calculating sqrt using binary search
def sqrt(x):
    if x == 0 or x == 1:
        return x
    start = 1
    end = x
    while start <= end:
        mid = (start + end) // 2
        if mid * mid == x:
            return mid
        if mid * mid < x:
            start = mid + 1
            result = mid
        else:
            end = mid - 1
    return result

"""
The concept we can use here is that, we can build a cumulative sum of all prime numbers
till the maximum limit and then find the largest difference. Since anyways we want
consequtive numbers, we don't have to take the pain of dynamic programming which
is used to handle the case of deleting certain numbers only from the set.
"""


def func(lim):
    prefix_sum = [2]
    prime_list = [2]
    for num in range(3, lim, 2):
        if is_prime(num):
            prime_list.append(num)
            prefix_sum.append(prefix_sum[-1] + num)

    # print(prefix_sum[-1])

    max_length = 0
    max_prime = 0
    for i in range(len(prefix_sum)):
        for j in range(i - (max_length + 1), -1, -1):
            if prefix_sum[i] - prefix_sum[j] > lim:
                break
            if prefix_sum[i] - prefix_sum[j] in prime_list:
                max_length = i - j
                max_prime = prefix_sum[i] - prefix_sum[j]

    return max_prime


# Example usage
lim = 1000000
result = func(lim)
print(result)
```
